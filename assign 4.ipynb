```python
import psutil
import pandas as pd
import datetime
import time
import random

# Settings
duration_minutes = 720  # Adjust to create a large dataset
sampling_rate_hz = 10
num_samples = duration_minutes * 60 * sampling_rate_hz

# Initialize lists to store data
timestamps = []
cpu_temperatures = []
cpu_usages = []
cpu_loads = []
memory_usages = []
battery_levels = []
cpu_powers = []

# Collect data
for i in range(num_samples):
    try:
        # Get current time
        current_time = datetime.datetime.now()
        timestamps.append(current_time)
        
        # Generate synthetic CPU temperature data
        cpu_temp = random.uniform(30, 90)  # Normal CPU temperature range
        cpu_temperatures.append(cpu_temp)
        
        # Generate synthetic CPU power data
        cpu_power = random.uniform(10, 70)  # Normal CPU power range
        cpu_powers.append(cpu_power)
        
        # Get CPU usage
        cpu_usage = psutil.cpu_percent(interval=1/sampling_rate_hz)
        cpu_usages.append(cpu_usage)
        
        # Get CPU load (1 minute average)
        cpu_load = psutil.getloadavg()[0]
        cpu_loads.append(cpu_load)
        
        # Get memory usage
        memory_usage = psutil.virtual_memory().percent
        memory_usages.append(memory_usage)
        
        # Generate synthetic battery level data
        battery_level = random.uniform(20, 100)  # Normal battery level range
        battery_levels.append(battery_level)
        
        # Introduce anomalies randomly (e.g., 10% chance)
        if random.random() < 0.1:  # 10% chance to introduce anomaly
            cpu_usages[-1] = random.uniform(90, 100)  # High CPU usage
        if random.random() < 0.1:
            cpu_temperatures[-1] = random.uniform(90, 105)  # High temperature
        if random.random() < 0.1:
            memory_usages[-1] = random.uniform(95, 100)  # High memory usage
        if random.random() < 0.1:
            battery_levels[-1] = random.uniform(0, 10)  # Low battery level
        if random.random() < 0.1:
            cpu_powers[-1] = random.uniform(50, 100)  # Unusually high CPU power

    except Exception as e:
        print(f"Error collecting data: {e}")
        cpu_temperatures.append(None)
        cpu_usages.append(None)
        cpu_loads.append(None)
        memory_usages.append(None)
        battery_levels.append(None)
        cpu_powers.append(None)

# Create DataFrame
data = {
    'timestamp': timestamps,
    'cpu_temperature': cpu_temperatures,
    'cpu_usage': cpu_usages,
    'cpu_load': cpu_loads,
    'memory_usage': memory_usages,
    'battery_level': battery_levels,
    'cpu_power': cpu_powers
}
df_real = pd.DataFrame(data)

# Save to CSV
df_real.to_csv('hardware_monitor_data.csv', index=False)

# Display the first few rows of the DataFrame
print(df_real.head())
```

```python
import psutil
import pandas as pd
import datetime
import random

# Settings for 1GB Dataset
target_file_size_gb = 1  # Desired size in GB
current_file_size_gb = 0.04  # Current size
scaling_factor = int(target_file_size_gb / current_file_size_gb)

# Adjust the duration to generate more data
duration_minutes = 720 * scaling_factor  # Scale up duration
sampling_rate_hz = 10
num_samples = duration_minutes * 60 * sampling_rate_hz

# Initialize lists to store data
timestamps = []
cpu_temperatures = []
cpu_usages = []
cpu_loads = []
memory_usages = []
battery_levels = []
cpu_powers = []

print(f"Generating {num_samples} samples to approximate 1GB of data...")

# Collect data
for i in range(num_samples):
    try:
        # Get current time
        current_time = datetime.datetime.now()
        timestamps.append(current_time)
        
        # Generate synthetic CPU temperature data
        cpu_temp = random.uniform(30, 90)  # Normal CPU temperature range
        cpu_temperatures.append(cpu_temp)
        
        # Generate synthetic CPU power data
        cpu_power = random.uniform(10, 70)  # Normal CPU power range
        cpu_powers.append(cpu_power)
        
        # Get CPU usage
        cpu_usage = random.uniform(0, 100)  # Simulate CPU usage
        cpu_usages.append(cpu_usage)
        
        # Get CPU load (1 minute average)
        cpu_load = random.uniform(0, 10)  # Simulate CPU load
        cpu_loads.append(cpu_load)
        
        # Get memory usage
        memory_usage = random.uniform(0, 100)  # Simulate memory usage
        memory_usages.append(memory_usage)
        
        # Generate synthetic battery level data
        battery_level = random.uniform(20, 100)  # Normal battery level range
        battery_levels.append(battery_level)
        
        # Introduce anomalies randomly (e.g., 10% chance)
        if random.random() < 0.1:  # 10% chance to introduce anomaly
            cpu_usages[-1] = random.uniform(90, 100)  # High CPU usage
        if random.random() < 0.1:
            cpu_temperatures[-1] = random.uniform(90, 105)  # High temperature
        if random.random() < 0.1:
            memory_usages[-1] = random.uniform(95, 100)  # High memory usage
        if random.random() < 0.1:
            battery_levels[-1] = random.uniform(0, 10)  # Low battery level
        if random.random() < 0.1:
            cpu_powers[-1] = random.uniform(50, 100)  # Unusually high CPU power

    except Exception as e:
        print(f"Error collecting data: {e}")
        cpu_temperatures.append(None)
        cpu_usages.append(None)
        cpu_loads.append(None)
        memory_usages.append(None)
        battery_levels.append(None)
        cpu_powers.append(None)

# Create DataFrame
data = {
    'timestamp': timestamps,
    'cpu_temperature': cpu_temperatures,
    'cpu_usage': cpu_usages,
    'cpu_load': cpu_loads,
    'memory_usage': memory_usages,
    'battery_level': battery_levels,
    'cpu_power': cpu_powers
}
df_real = pd.DataFrame(data)

# Save to CSV
df_real.to_csv('hardware_monitor_data_1gb.csv', index=False)

# Display the first few rows of the DataFrame
print(df_real.head())
```

```python
import pandas as pd
import numpy as np
import datetime
import random
# Introduce anomalies at random indices
anomaly_indices = np.random.choice(num_samples, size=int(0.01 * num_samples), replace=False)  # 1% anomalies

# Apply anomalies to the data
df_real.loc[anomaly_indices, "cpu_temperature"] = np.random.uniform(100, 180, size=len(anomaly_indices))
df_real.loc[anomaly_indices, "cpu_usage"] = np.random.uniform(90, 100, size=len(anomaly_indices))
df_real.loc[anomaly_indices, "memory_usage"] = np.random.uniform(95, 100, size=len(anomaly_indices))

# Save the data with anomalies
df_real.to_csv("synthetic_time_series_with_anomalies.csv", index=False)
print("Anomalous data saved to 'synthetic_time_series_with_anomalies.csv'")
```

```python
# Optimized Anomaly Detection Using Vectorization
import numpy as np
import pandas as pd

# Load the data
df_anomalous = pd.read_csv("synthetic_time_series_with_anomalies.csv")

# Function to compute Z-scores in a vectorized way
def detect_anomalies(data, threshold=3):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return np.abs(z_scores) > threshold

# Detect anomalies using vectorized operations
df_anomalous["temperature_anomaly"] = detect_anomalies(df_anomalous["cpu_temperature"])
df_anomalous["usage_anomaly"] = detect_anomalies(df_anomalous["cpu_usage"])
df_anomalous["memory_anomaly"] = detect_anomalies(df_anomalous["memory_usage"])

# Combine anomaly flags into a single column
df_anomalous["overall_anomaly"] = (
    df_anomalous["temperature_anomaly"] |
    df_anomalous["usage_anomaly"] |
    df_anomalous["memory_anomaly"]
)

# Save the anomaly detection results
df_anomalous.to_csv("time_series_with_detected_anomalies.csv", index=False)
print("Optimized anomaly detection results saved to 'time_series_with_detected_anomalies.csv'")
```

```python
import matplotlib.pyplot as plt

# Filter data for visualization (first 1000 rows only)
visualization_data = df_anomalous.head(1000)

# Create subplots for visualization
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# CPU Temperature Plot
axs[0].plot(visualization_data.index, visualization_data["cpu_temperature"], label="Data", color="blue")
axs[0].scatter(
    visualization_data.index[visualization_data["temperature_anomaly"]],
    visualization_data["cpu_temperature"][visualization_data["temperature_anomaly"]],
    color="red",
    label="Anomalies"
)
axs[0].set_title("CPU Temperature with Anomalies Highlighted")
axs[0].set_ylabel("CPU Temperature")
axs[0].legend()

# CPU Usage Plot
axs[1].plot(visualization_data.index, visualization_data["cpu_usage"], label="Data", color="blue")
axs[1].scatter(
    visualization_data.index[visualization_data["usage_anomaly"]],
    visualization_data["cpu_usage"][visualization_data["usage_anomaly"]],
    color="red",
    label="Anomalies"
)
axs[1].set_title("CPU Usage with Anomalies Highlighted")
axs[1].set_ylabel("CPU Usage (%)")
axs[1].legend()

# Memory Usage Plot
axs[2].plot(visualization_data.index, visualization_data["memory_usage"], label="Data", color="blue")
axs[2].scatter(
    visualization_data.index[visualization_data["memory_anomaly"]],
    visualization_data["memory_usage"][visualization_data["memory_anomaly"]],
    color="red",
    label="Anomalies"
)
axs[2].set_title("Memory Usage with Anomalies Highlighted")
axs[2].set_xlabel("Index")
axs[2].set_ylabel("Memory Usage (%)")
axs[2].legend()

plt.tight_layout()
plt.show()
```
